# Basics

```python
import qiskit as qk
import numpy as np
```

## Hadamard gate circuit

```python
circuit = qk.QuantumCircuit(2,2)
circuit.h(0)
circuit.cx(0, 1)
circuit.measure([0,1],[0,1])
circuit.draw(output="mpl")
```

## Not gate circuit

```python
circuit = qk.QuantumCircuit(2,2)
circuit.x(0)
circuit.cx(0, 1)
circuit.measure([0,1],[0,1])
circuit.draw(output="mpl")
```
